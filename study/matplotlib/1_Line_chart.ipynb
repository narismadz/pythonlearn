```python
import matplotlib.pyplot as plt # import graph module
```

```python
# first line graph [x , y]
plt.plot([4, -5, 9, 16], [1,6,-7, 6])
plt.show()
```

```python
# plot as marker "o" with green color
plt.plot([4, -5, 9, 16], [1,6,-7, 6], "go") # 'g = green' 'o = dot'
plt.show()
```

```python
# plot as marker "o" with green color and red line against black line and pink line
plt.plot([4, -5, 9, 16], [1,6,-7, 6], "go-")
plt.plot([4, 5, -3, 8], [-1,14,-7, 10], "ko-")
plt.plot([2, -8,12, -12], [-1,14,-7, 10], color = "pink", marker="o")
plt.show()
```

```python
# brown dotted line
plt.plot([1,3,2],[9,8,13], color = "brown", linestyle="--" )
plt.show()
```

```python
# Function plot method with incremental plot from x = 2014-2021
def plot1():
    x = range(2014, 2021)
    y = [20, 25, 18, 15, 17, 20, 22]
    plt.plot(x, y, color = "red")
    plt.show()
```

```python
#plot with actual and estimation
def plot2():
    x = range(2014, 2021)
    y = [20, 25, 18, 15, 17, 20, 22]
    plt.plot(x[0:4], y[0:4], color="deepskyblue", label="actual") #plot only first 4 years
    plt.plot(x[3:], y[3:], color="deepskyblue", linestyle="--", label="forecast") #estimate starts
    #add label and legend
    plt.title("sales forecast")
    plt.ylabel("sales ('000)")
    plt.xlabel("year")
    plt.legend()
    plt.show()
```

# 2. Bar chart

```python
import matplotlib.pyplot as plt
```

```python
#verticle bar as default

def demo1():
    x = range(5)          # will assumed 0 to 5
    y = [20,25,35,21,36]
    xtick = ('p1', 'p2', 'p3', 'p4', 'p5')
    plt.bar(x,y)          # will get default color
    plt.show()
```

```python
# replace x with label xtick
# insert title
# insert y label
# insert avg as dotted line along with x axis

def demo2():
    x = range(5)          # will assumed 0 to 5
    y = [20,25,35,21,36]
    xtick = ('p1', 'p2', 'p3', 'p4', 'p5') #label xtick
    avg = sum(y)/5
    print('avg =',avg)
    
    plt.bar(x,y, color = "Darkred")
    plt.xticks(x,xtick) 
    plt.title('Products vs Prices \n FY19H2', color='Darkgreen', fontsize = '17') # title
    plt.ylabel('Price') # y label
    plt.axhline(avg, color = 'G', linestyle = '--', label = 'Average') # avg horizintal dotted line
    plt.legend() # show avg legend
    plt.show()
```

```python
# run both functions
if __name__ == '__main__':
    demo1()
    demo2()
```

```python
# horizontal graph --> plt.barh

def demo3():
    x = range(5)          # will assumed 0 to 5
    y = [20,25,35,21,36]
    xtick = ('p1', 'p2', 'p3', 'p4', 'p5') #label xtick
    avg = sum(y)/5
        
    plt.barh(x,y, color = "grey")
    plt.yticks(x,xtick) 
    plt.title('Products vs Prices \n FY19H2', color='Darkgreen', fontsize = '17') # title
    plt.xlabel('Price') # x label
    plt.axvline(avg, color = 'Green', linestyle = ':', linewidth = 4, label = 'Average') # avg vertical dotted line 
    plt.legend() # show avg legend
    plt.show()
```

```python
# run functions
if __name__ == '__main__':
    demo3() 
```
